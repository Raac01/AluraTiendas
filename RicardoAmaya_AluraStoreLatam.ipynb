### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 2. Ventas por categoría

```python
def ventas_por_categoria(df, nombre_tienda):
    resumen = df.groupby('Categoría del Producto')['Precio'].sum().reset_index()
    resumen.columns = ['Categoría del Producto', f'Ventas {nombre_tienda}']
    return resumen

# Obtener ventas por categoría para cada tienda
ventas1 = ventas_por_categoria(tienda, 'Tienda 1')
ventas2 = ventas_por_categoria(tienda2, 'Tienda 2')
ventas3 = ventas_por_categoria(tienda3, 'Tienda 3')
ventas4 = ventas_por_categoria(tienda4, 'Tienda 4')

# Unir los resultados por categoría
ventas_totales = ventas1.merge(ventas2, on='Categoría del Producto', how='outer') \
                         .merge(ventas3, on='Categoría del Producto', how='outer') \
                         .merge(ventas4, on='Categoría del Producto', how='outer')

ventas_totales = ventas_totales.fillna(0)
pd.set_option('display.expand_frame_repr', False)


# Mostrar el resultado
print(ventas_totales)
```

# 3. Calificación promedio de la tienda


```python
# Función para calcular promedio de calificación por categoría
def promedio_calificacion(df, nombre_tienda):
    resumen = df.groupby('Categoría del Producto')['Calificación'].mean().reset_index()
    resumen.columns = ['Categoría del Producto', f'Calificación Prom. {nombre_tienda}']
    return resumen

# Obtener promedios por categoría para cada tienda
prom1 = promedio_calificacion(tienda, 'Tienda 1')
prom2 = promedio_calificacion(tienda2, 'Tienda 2')
prom3 = promedio_calificacion(tienda3, 'Tienda 3')
prom4 = promedio_calificacion(tienda4, 'Tienda 4')

# Unir los resultados
promedios_totales = prom1.merge(prom2, on='Categoría del Producto', how='outer').merge(prom3, on='Categoría del Producto', how='outer').merge(prom4, on='Categoría del Producto', how='outer')

promedios_totales = promedios_totales.fillna(0)
pd.set_option('display.expand_frame_repr', False)


# Mostrar el resultado
print(promedios_totales)
```

# 4. Productos más y menos vendidos

```python
# Unir todos los DataFrames en uno solo
df_total = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
# Sumar ventas por producto
ventas_por_producto = df_total.groupby('Producto')['Precio'].sum().reset_index()

# Ordenar de mayor a menor
productos_mas_vendidos = ventas_por_producto.sort_values(by='Precio', ascending=False)

print("Productos más vendidos:")
print(productos_mas_vendidos.head(10))
```

```python
print("Productos menos vendidos:")
print(productos_mas_vendidos.tail(10))
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio por tienda
envio1 = tienda['Costo de envío'].mean()
envio2 = tienda2['Costo de envío'].mean()
envio3 = tienda3['Costo de envío'].mean()
envio4 = tienda4['Costo de envío'].mean()

# Crear resumen como DataFrame
envios_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo de Envío Promedio': [envio1, envio2, envio3, envio4]
})

print(envios_promedio)
```
